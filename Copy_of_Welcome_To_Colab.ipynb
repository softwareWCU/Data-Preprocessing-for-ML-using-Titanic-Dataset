## Handle Missing Values

### Subtask:
Address missing values in the 'Age' and 'Fare' columns by imputing them with the median of their respective columns. For the 'Embarked' column, impute missing values with the mode (most frequent value).


**Reasoning**:
First, I will calculate the median of the 'Age' column from the DataFrame `df` and store it in a variable named `median_age` as per the first instruction.



````
median_age = df['Age'].median()
print(f"Median age: {median_age}")
````

**Reasoning**:
Now, I will fill the missing values in the 'Age' column of `df` with the previously calculated `median_age`, as per the second instruction.



````
df['Age'].fillna(median_age, inplace=True)
print("Missing values in 'Age' column after imputation:")
print(df['Age'].isnull().sum())
````

**Reasoning**:
The previous step successfully filled missing 'Age' values but triggered a `FutureWarning` about modifying a view. To avoid this and continue with the next instructions (imputing 'Fare' with its median and 'Embarked' with its mode), I will explicitly create a copy of the DataFrame to ensure modifications are made on the copy, and then apply all remaining imputation steps in a single code block to complete the subtask.



````
df_copy = df.copy()

# Impute missing 'Age' values with the median (addressing Future Warning)
median_age = df_copy['Age'].median()
df_copy['Age'] = df_copy['Age'].fillna(median_age)
print(f"Median age: {median_age}")
print(f"Missing values in 'Age' column after imputation: {df_copy['Age'].isnull().sum()}")

# Calculate median for 'Fare' and impute missing values
median_fare = df_copy['Fare'].median()
df_copy['Fare'] = df_copy['Fare'].fillna(median_fare)
print(f"Median fare: {median_fare}")
print(f"Missing values in 'Fare' column after imputation: {df_copy['Fare'].isnull().sum()}")

# Calculate mode for 'Embarked' and impute missing values
mode_embarked = df_copy['Embarked'].mode()[0]
df_copy['Embarked'] = df_copy['Embarked'].fillna(mode_embarked)
print(f"Mode for Embarked: {mode_embarked}")
print(f"Missing values in 'Embarked' column after imputation: {df_copy['Embarked'].isnull().sum()}")

# Verify all missing values in relevant columns
print("\nMissing values after all imputations:")
print(df_copy[['Age', 'Fare', 'Embarked']].isnull().sum())
````

## Encode Categorical Features

### Subtask:
Convert categorical features 'Sex' and 'Pclass' into a numerical format suitable for machine learning models using one-hot encoding. This will create new columns for each category and drop the original categorical columns to avoid multicollinearity.


**Reasoning**:
I will perform one-hot encoding on the specified categorical columns and then drop the original columns along with irrelevant ones to prepare the DataFrame for machine learning. I will print the first 5 rows and info of the new DataFrame to confirm the changes.



````
df_encoded = pd.get_dummies(df_copy, columns=['Sex', 'Pclass', 'Embarked'], drop_first=True)
df_final = df_encoded.drop(columns=['Name', 'Ticket', 'PassengerId'])

print(df_final.head())
print(df_final.info())
````

**Reasoning**:
The previous one-hot encoding resulted in duplicate columns due to inconsistent casing in the original categorical features. I need to standardize the casing of 'Sex', 'Pclass', and 'Embarked' columns first, then re-apply one-hot encoding. Additionally, the 'Cabin' column, which has a large number of missing values and is not directly relevant to the current task of handling specified missing values and encoding, should be dropped. I will also drop 'Name', 'Ticket', and 'PassengerId' as they are not needed for machine learning, then display the head and info of the final DataFrame to verify the changes.



````
df_cleaned_categories = df_copy.copy()

# Standardize casing for categorical columns to avoid duplicate dummy variables
df_cleaned_categories['Sex'] = df_cleaned_categories['Sex'].str.lower()
df_cleaned_categories['Pclass'] = df_cleaned_categories['Pclass'].astype(str).str.lower()
df_cleaned_categories['Embarked'] = df_cleaned_categories['Embarked'].str.lower()

# Perform one-hot encoding on 'Sex', 'Pclass', and 'Embarked' columns
df_encoded = pd.get_dummies(df_cleaned_categories, columns=['Sex', 'Pclass', 'Embarked'], drop_first=True)

# Drop original categorical columns and other irrelevant columns
df_final = df_encoded.drop(columns=['Name', 'Ticket', 'PassengerId', 'Cabin'])

print("DataFrame after cleaning categories, one-hot encoding, and dropping irrelevant columns:")
print(df_final.head())
print(df_final.info())
````

**Reasoning**:
The previous one-hot encoding still produced duplicate-like columns (e.g., `Sex_male`, `Sex_male` and `Embarked_c`, `Embarked_c`), indicating issues with unique values or casing even after initial standardization. To properly apply one-hot encoding and avoid redundant dummy variables, I will first inspect the unique values of the 'Sex', 'Pclass', and 'Embarked' columns in the `df_cleaned_categories` DataFrame. This will help identify any remaining inconsistencies before re-encoding and ensure that `drop_first=True` works as expected.



````
print("Unique values in 'Sex' after standardizing casing:")
print(df_cleaned_categories['Sex'].unique())

print("\nUnique values in 'Pclass' after standardizing casing:")
print(df_cleaned_categories['Pclass'].unique())

print("\nUnique values in 'Embarked' after standardizing casing:")
print(df_cleaned_categories['Embarked'].unique())
````

**Reasoning**:
The previous inspection revealed inconsistencies like trailing/leading spaces and different representations of the same category (e.g., 'male' vs 'male ', '3' vs '3rd ', 'c' vs 'c '). To ensure proper one-hot encoding without duplicate dummy variables, I will clean these inconsistencies by stripping whitespace and standardizing the values in 'Sex', 'Pclass', and 'Embarked' columns. Then, I will re-apply one-hot encoding, drop irrelevant columns, and verify the final DataFrame.



````
df_cleaned_categories = df_copy.copy()

# Clean 'Sex' column: strip whitespace
df_cleaned_categories['Sex'] = df_cleaned_categories['Sex'].str.lower().str.strip()

# Clean 'Pclass' column: strip whitespace and standardize '3rd' to '3'
df_cleaned_categories['Pclass'] = df_cleaned_categories['Pclass'].astype(str).str.lower().str.strip()
df_cleaned_categories['Pclass'] = df_cleaned_categories['Pclass'].replace('3rd', '3')

# Clean 'Embarked' column: strip whitespace and standardize full words to single chars
df_cleaned_categories['Embarked'] = df_cleaned_categories['Embarked'].str.lower().str.strip()
df_cleaned_categories['Embarked'] = df_cleaned_categories['Embarked'].replace({'cherbourg': 'c', 'southampton': 's'})

# Verify unique values after cleaning
print("Unique values in 'Sex' after cleaning:")
print(df_cleaned_categories['Sex'].unique())
print("\nUnique values in 'Pclass' after cleaning:")
print(df_cleaned_categories['Pclass'].unique())
print("\nUnique values in 'Embarked' after cleaning:")
print(df_cleaned_categories['Embarked'].unique())

# Perform one-hot encoding on 'Sex', 'Pclass', and 'Embarked' columns
df_encoded = pd.get_dummies(df_cleaned_categories, columns=['Sex', 'Pclass', 'Embarked'], drop_first=True, dtype=int)

# Drop original categorical columns and other irrelevant columns
df_final = df_encoded.drop(columns=['Name', 'Ticket', 'PassengerId', 'Cabin'])

print("\nDataFrame after cleaning categories, one-hot encoding, and dropping irrelevant columns:")
print(df_final.head())
print(df_final.info())
print(df_final.columns)
````

## Analyze Categorical Features

### Subtask:
Generate bar plots for categorical features (one-hot encoded 'Sex', 'Pclass', and 'Embarked') to visualize their counts and distribution.


**Reasoning**:
I will generate bar plots for the identified one-hot encoded categorical features to visualize their counts and distributions, as per the instructions.

### Numerical Feature Distributions:
*   **Age**: The age distribution is slightly right-skewed, with a significant portion of passengers in the 20-40 age range. There are also peaks at younger ages, indicating children.
*   **Fare**: The fare distribution is highly right-skewed, with most passengers paying lower fares and a few paying very high fares. This suggests a wide range of economic backgrounds among the passengers.
*   **SibSp (Siblings/Spouses Aboard)**: A large majority of passengers traveled without siblings or spouses (SibSp = 0). As the number of siblings/spouses increases, the count of passengers decreases sharply.
*   **Parch (Parents/Children Aboard)**: Similar to SibSp, most passengers traveled without parents or children (Parch = 0), with counts decreasing significantly for higher values.

### Categorical Feature Distributions:
*   **Sex_male**: There are more male passengers than female passengers in the dataset.
*   **Pclass_2 (2nd Class)**: A smaller proportion of passengers were in 2nd class compared to 3rd class.
*   **Pclass_3 (3rd Class)**: A significant portion of passengers were in 3rd class.
*   **Embarked_q (Queenstown)**: The fewest passengers embarked from Queenstown.
*   **Embarked_s (Southampton)**: The majority of passengers embarked from Southampton.

### Survival by Features:
*   **Survival by Sex**: A higher proportion of females survived compared to males. This is a strong indicator of the "women and children first" policy.
*   **Survival by Pclass**: Passengers in 1st class (`Pclass_1` which is implicitly the inverse of `Pclass_2` and `Pclass_3` being `0`) had a higher survival rate than those in 2nd (`Pclass_2`) and 3rd class (`Pclass_3`). 3rd class passengers had the lowest survival rate, indicating socioeconomic status played a significant role.
*   **Survival by Embarked**: Passengers embarking from Cherbourg (implicitly `Embarked_c` where both `Embarked_q` and `Embarked_s` are `0`) appear to have a slightly higher survival rate compared to those from Queenstown (`Embarked_q`) and Southampton (`Embarked_s`).
*   **Survival by Age**: The age distribution shows that younger children (0-10 years old) had a relatively higher survival rate. There's also a noticeable peak in survival for young adults, while older adults (above 60) had a lower survival rate. The highest number of casualties were among young adults (20-40 years old).
